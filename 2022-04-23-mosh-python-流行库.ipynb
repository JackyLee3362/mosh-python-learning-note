## Numpy `1:06:38`

```python
import numpy as np

array = np.array([1,2,3])
print(array)
print(type(array))

array = np.array([ [1,2,3], [4,5,6]])
print(array.shape) # 返回长度

np.zeros((3,4),dtype=int) # 返回一个3*4大小的全零(浮点型）矩阵
np.zeros((3,4),dtype=int) # 返回一个3*4大小的全零（整形）矩阵
np.ones((3,4))
np.full((3,4),5)
array = np.random.random((3,4))
print(array)
print(array[0,0])  # 返回第一行第一列，或者写作array[0][0]
print(array > 0.2) # 返回布尔矩阵
print(array[array > 0.2])
print(np.sum(array))
print(np.floor(array))  # 或者ceil， round
```
